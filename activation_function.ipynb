```python
import torch
import d2l.torch as d2l
```

## ReLU函数

```python
x = torch.arange(-8.0,8.0,0.1,requires_grad=True)
y = torch.relu(x)
y.sum().backward()
d2l.plot(x.detach().numpy(),[y.detach().numpy(), x.grad.detach().numpy()], 'x', 'y',legend=['y','dy/dx'], figsize=(5,2.5))
```

## sigmoid函数

```python
x = torch.arange(-8.0,8.0,0.1,requires_grad=True)
y = torch.sigmoid(x)
y.sum().backward()
d2l.plot(x.detach().numpy(),[y.detach().numpy(), x.grad.detach().numpy()], 'x', 'y',legend=['y','dy/dx'], figsize=(5,2.5))
```

## tanh函数

```python
x = torch.arange(-8.0,8.0,0.1,requires_grad=True)
y = torch.tanh(x)
y.sum().backward()
d2l.plot(x.detach().numpy(),[y.detach().numpy(), x.grad.detach().numpy()], 'x', 'y',legend=['y','dy/dx'], figsize=(5,2.5))
```
